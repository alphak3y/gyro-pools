```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import seaborn as sns
sns.set_theme()
```

```python
fsingle = pd.read_feather("../data/errors_single_decimal.feather")
fdouble = pd.read_feather("../data/errors_double_decimal.feather")
```

```python
fsingle_errs = fsingle[fsingle['error'] > 0]
fdouble_errs = fdouble[fdouble['error'] > 0]
```

```python
len(fsingle_errs), len(fdouble_errs)
```

```python
np.log10(fsingle_errs).max()
```

```python
np.log10(fdouble_errs).max()
```

```python
sns.distplot(np.log10(fsingle_errs))
plt.title("Single Precision")
plt.xlabel("log10(loss in assets)")
plt.show()
```

```python
sns.distplot(np.log10(fdouble_errs))
plt.title("Double Precision")
plt.xlabel("log10(loss in assets)")
plt.show()
```
