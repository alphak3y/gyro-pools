```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

import seaborn as sns
sns.set_theme()
```

```python
fsingle = pd.read_feather("../data/errors_single_decimal.feather")
fdouble = pd.read_feather("../data/errors_double_decimal.feather")
fsolidity = pd.read_feather("../data/errors_solidity.feather")
```

```python
fsingle_errs = fsingle[fsingle['error'] > 0]
fdouble_errs = fdouble[fdouble['error'] > 0]
fsolidity_errs = fsolidity[fsolidity.loss_ub_sol > 0]
```

```python
len(fsingle_errs), len(fdouble_errs), len(fsolidity_errs)
```

```python
len(fsingle_errs)/len(fsingle), len(fdouble_errs)/len(fdouble), len(fsolidity_errs)/len(fsolidity)
```

```python
np.log10(fsingle_errs).max()
```

```python
np.log10(fdouble_errs).max()
```

```python
(fsolidity.loss_ub > 0).sum()
```

```python
np.log10(fsolidity_errs).max()
```

```python
sns.distplot(np.log10(fsingle_errs))
plt.title("Single Precision")
plt.xlabel("log10(loss in assets)")
plt.show()
```

```python
sns.distplot(np.log10(fdouble_errs))
plt.title("Double Precision")
plt.xlabel("log10(loss in assets)")
plt.show()
```

```python
s = fsolidity.loss_ub
sns.distplot(np.log10(s[s > 0]))
plt.title("Single (new data)")
plt.xlabel("log10(loss in assets)")
plt.show()
```

```python
sns.distplot(np.log10(fsolidity_errs.loss_ub_sol))
plt.title("Solidity")
plt.xlabel("log10(loss in assets)")
plt.show()
```
